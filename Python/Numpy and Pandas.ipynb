**Data Cleaning**

```python
import pandas as pd
import numpy as np
#creating a dataframe using dict
df = pd.DataFrame(data={
    'Temperature':[25,np.nan,34,20,15,19,18,np.nan,25],
    'Humidity':[20,20,12,np.nan,13,15,np.nan,np.nan,20],
    'Situation':['Humid',np.nan, 'Too-Sunny','cool','Too Cool','Normal','Normal',np.nan,'Humid']
})
df
```

```python
#checking the dataframe for a missing value
df.isna()
```

```python
df.isna().sum()
```

```python
#using visual approach to check the missing values
import seaborn as sns
sns.heatmap(df.isnull())
```

```python
#Deleting missing values anywhere is table
tempdf = df.dropna()
tempdf
```

```python
#Deleting record which have all missing values
tempdf = df.dropna(how="all")
tempdf
```

```python
#Filling using forward method - Top Down approach
ffill = df.fillna(method="ffill")
ffill
```

```python
#Filling using backward method - Bottom Top approach
dfill = df.fillna(method="dfill")
dfill
```

```python
sns.heatmap(ffill.isnull)
```

```python
#df is my main dataframe
#ffill is the dataframe for forward method
# tempdf is the dataframe for removing whole record with missing values, NaN
```

```python
tempdf
```

```python
meanv = tempdf["Temperature"].mean()
meanv
```

```python
#to fill and make it parmanent change in the table i.e. tempdf using mean value and inplace = True
tempdf["Temperature"].fillna(meanv, inplace=True)
```

```python
tempdf
```

```python
meanh = tempdf["Temperature"].mean()
meanh
tempdf["Temperature"].fillna(meanh, inplace=True)
```

```python
tempdf
```

```python
tempdf.drop_duplicates()
```

```python
from matplotlib import pyplot as plt
x = [0,1,2,3,4]
y = [0,10,20,30,40]
plt.plot(x,y)
plt.title("Simple Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
#Highlighting the intersection points
from matplotlib import pyplot as plt
x = [0,1,2,3,4]
y = [0,10,20,30,40]
plt.plot(x,y, marker = "h")
plt.title("Simple Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
#Styling or heading to plot
from matplotlib import pyplot as plt
x = [0,1,2,3,4]
y = [0,10,20,30,40]
plt.plot(x,y, marker = "h", ms=15)
f1 = {"family":"calibri","color":"red","size":"25"}
f2 = {"family":"calibri","color":"green","size":"14"}
plt.title("Simple Plot", fontdict=f1)
plt.xlabel("X-Axis", fontdict=f2)
plt.ylabel("Y-Axis", fontdict=f2)
```

```python
from matplotlib import pyplot as plt
x = [0,1,2,3,4]
y = [0,10,20,30,40]
plt.plot(x,y, marker = "h", ms=15)
f1 = {"family":"calibri","color":"red","size":"25"}
f2 = {"family":"calibri","color":"green","size":"14"}
plt.title("Simple Plot", fontdict=f1)
plt.xlabel("X-Axis", fontdict=f2)
plt.ylabel("Y-Axis", fontdict=f2)
plt.show()
```

```python
#Vertical subplot
from matplotlib import pyplot as plt
x = [0,1,2,3,4]
y = [0,10,20,30,40]
y1 = [30,20,40,10,50]
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("Sub Plot1")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.subplot(1,2,2)
plt.plot(x,y1)
plt.title("Sub Plot2")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
#Horizontal subplot
from matplotlib import pyplot as plt
x = [0,1,2,3,4]
y = [0,10,20,30,40]
y1 = [30,20,40,10,50]
plt.subplot(2,1,1)
plt.plot(x,y)
plt.title("Sub Plot1")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.subplot(2,1,2)
plt.plot(x,y1)
plt.title("Sub Plot2")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```
