```python
#FIZZBUZZ code
for n in range(1,1001):
    if n%3==0 and n%5==0:
        print("FIZZBUZZ")
    elif n%3==0:
        print("FIZZ")
    elif n%5==0:
        print("BUZZ")
    else:
        print(n)
```

```python
i = 1
maxnum = 1000
while (i <= maxnum):
    if i%3==0 and i%5==0:
        print("FIZZBUZZ")
    elif i%3==0:
        print("FIZZ")
    elif i%5==0:
        print("BUZZ")
    else:
        print(i)

    i = i+1
```
